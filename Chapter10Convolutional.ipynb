# Multi-layer Neural Network

recap_ffn.png

```python
import tensorflow as tf
```

```python
model = tf.keras.Sequential([
  tf.keras.layers.Dense(1000, activation="elu"),
  tf.keras.layers.Dense(500, activation="elu"),
  tf.keras.layers.Dense(10, activation="softmax")
])
```

```python
model(tf.zeros([1, 784]))
```

```python
model.summary()
```

# Convolutional Neural Network

our_first_convnet.png

```python
model = tf.keras.Sequential([
  tf.keras.layers.Conv2D(filters=4, kernel_size=3, strides=(1, 1), padding='same', activation="relu"),
  tf.keras.layers.MaxPool2D(pool_size=(2, 2), strides=None, padding='valid'),
  tf.keras.layers.Conv2D(filters=8, kernel_size=3, strides=(1, 1), padding='same', activation="relu"),
  tf.keras.layers.MaxPool2D(pool_size=(2, 2), strides=None, padding='valid'),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(20, activation="relu"),
  tf.keras.layers.Dense(10, activation="softmax")
])
```

```python
model(tf.zeros([1, 28, 28, 1]))
```

```python
model.summary()
```
