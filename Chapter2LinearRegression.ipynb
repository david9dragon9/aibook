```python
w = 0
b = 0
```

```python
x = [1, 2, 3]
y = [1, 1.4, 2.1]
```

```python
def mse(x, y, w, b):
    error = 0
    for i in range(len(x)):
        error += (y[i] - (w*x[i] + b))**2
    error /= len(x)
    return error
```

```python
def gradient(x, y, w, b):
    dw = 0
    db = 0
    for i in range(len(x)):
        dw += 2 * (y[i] - (w*x[i] + b)) * -x[i]
        db += 2 * (y[i] - (w*x[i] + b)) * -1
    dw /= len(x)
    db /= len(x)
    return dw, db
```

```python
iterations = 1000
learning_rate = 0.1
for i in range(iterations):
    dw, db = gradient(x, y, w, b)
    print("current mse =", mse(x, y, w, b))
    w = w - learning_rate * dw
    b = b - learning_rate * db
```
